```python
def cost(X, y, theta):
    k = X.shape[0]
    total_cost = 0
    for i in range(k):
        total_cost += 1/k * (y[i] - ( theta.dot(X[i,:]) )**2)
    return total_cost
                             
```

```python
def step_gradient(X, y, learning_rate, theta):
    k = X.shape[0]
    n = X.shape[1]
    gradients = np.zeros(n)
    for i in range(k):
        for j in range(n):
            gradients[j] += (-1/k) * ( y[i] - (theta.dot(X[i,:])) ) * X[i,j]
    theta = theta - learning_rate * gradients
    return theta
```

```python
def gradient_descent(X, y, learning_rate, iterations):
    k = X.shape[0]
    n = X.shape[1]
    theta = np.zeros(n)                #random initialization
    for i in range(iterations):
        theta = step_gradient(X, y, learning_rate, theta)
        print(i, 'cost:', cost(X, y, theta))
    print(theta)
    
    return theta
```
